```python
n_rows=10
n_clos=1
cols=('usia',)
df=pd.DataFrame(np.random.randint(1,99, size=(n_rows,n_clos)),
                columns=cols)
df
```

```python
df['kelompok_usia'] = pd.cut(df['usia'],
                             bins = [0, 18, 65, 99],
                             labels = ['anak','dewasa','manula'])
df
```

```python
n_rows=5
n_clos=5
cols=tuple('ABCDE')
df=pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_clos)),
                columns=cols)
df
```

```python
df1=df.copy(deep=True)
df1=df1.drop([1,4])
df1
```

```python
df2=df.copy(deep=True)
df2=df2.drop([0,3])
df2
```

```python
df_inner=pd.merge(df1,df2,how='inner')
df_inner
```

```python
df_outer=pd.merge(df1,df2,how='outer')
df_outer
```
