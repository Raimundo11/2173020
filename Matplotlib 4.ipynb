```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np 
print(matplotlib.__version__)
print(np.__version__)
```

```python
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data)
plt.grid()
plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
plt.bar(kategori, 
        data,
        color='red',
        alpha=0.55)

plt.grid(linestyle='--',
         linewidth=2,
         axis='y',
         alpha=0.50)

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
plt.barh(kategori[::-1],data[::-1])

plt.grid(linestyle='--',
         linewidth=1,
         axis='x',
         alpha=0.50)

plt.xlabel('jumlah')
plt.ylabel('kategori')
plt.title('Contoh Bar Plot')
plt.show()
```

```python
data1=[25,85,75,40,60]
data2=[40,35,20,55,10]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label='Data 1')
plt.bar(kategori,data2,label='Data 2',bottom=data1)
plt.legend()

 
```

```python
data1=[25,85,75,40,60]
data2=[40,35,20,55,10]
kategori=['A','B','C','D','E']
x=np.arange(len(kategori))
x
```

```python
width=0.35

plt.bar(x-width/2,data1,width,label=data1)
plt.bar(x+width/2,data2,width,label=data2)
plt.xticks(x,kategori)

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.50)

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('Contoh Group Bar Plot')
plt.legend()
plt.show()
```

```python
kategori=['A','B','C','D','E']
data=[215,130,245,210,225]
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.1f%%',
       startangle=90)
plt.title('contoh pie plot ')
plt.show
```

```python
kategori=['A','B','C','D','E']
data=[215,130,245,210,225]
warna=['pink','cyan','#e03364','yellowgreen','skyblue']
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
       autopct='%1.1f%%',
       startangle=50)
plt.title('contoh pie plot ')
plt.legend()
plt.show()
```

```python
kategori=['A','B','C','D','E']
data=[215,130,245,210,225]
warna=['pink','cyan','#e03364','yellowgreen','skyblue']
explode_var=[0.,0.,0.105,0.,0.08]
```

```python
plt.pie(data,
       labels=kategori,
       colors=warna,
        explode=explode_var,
       autopct='%1.1f%%',
       startangle=50)
plt.title('contoh pie plot ')
plt.legend()
plt.show()
```
